**Question no. 1** Explain the difference between Relative, absolute and cell referencing?
**Answer** : **Relative cell referencing**: A cell address   thatchanges when copied to other cell. If we drag down cursor for other cell it changes automatically.
        ** Absolute cell referencing**: A cell address that not changes when we copied to other cell. We make a cell absolute by clicking F4. We make cell absolute because we want the cell referring with the other cell.
        ** Mixed cell referencing**: A cell address that mixed of relative and absolute cell. When we apply some formula to an specific cell we make some cell absolute and some cell relative, that we called mixed cell refernicing.
